## Create X-Y plots using a single line of code

```python
plot_xy([1,2,3,4,5],[3,4,5,6,7])
```

### Customize the plot with all matplotlib functionalities
* Labels - X axis, Y axis, data
* Change linewidths, linestyles, markers and markerstyles
* Plot title
* And much more...
For a complete list of options, run `print(plot_xy.__doc__)`

```python
plot_xy([1,2,3,4,5],[3,4,5,6,7], figsize=(6,6), label="Random Data 1", xlabel="Random X", ylabel="Random Y")
plot_xy([3,4,5],[6,7,10], new_fig=False, label="Random Data 2", lw=4, marker='o', ms=15)
plot_xy([1,2,3],[5,7,10], new_fig=False, lw=3, title='Plot X-Y')
```

## Plot Image Data with Options to Customize

```python
"""Generate random img data"""
import numpy as np
img_data = np.random.random(2500)
img_data = np.reshape(img_data, newshape=(50,50))
```

```python
"""Plot image data with scalebar in one line"""
plot_confocal(img_data, stepsize=0.1) # default unit is 'um' but can be changed
```

### Customize stepsize units, colormaps, colorbar labels and much more...
For a complete list of options, run `print(plot_confocal.__doc__)`

```python
plot_confocal(img_data, stepsize=0.5, units='nm', cmap="cividis", cbar_label="Height (nm)")
```
