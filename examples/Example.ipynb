## Create X-Y plots using a single line of code

```python
plot_xy([1,2,3]) # plots them as y-values, auto generates x-values
```

### Customize the plot with all matplotlib functionalities
* Labels - X axis, Y axis, data
* Change linewidths, linestyles, markers and markerstyles
* Plot title
* And much more...
For a complete list of options, run `print(plot_xy.__doc__)`

### Add plot title and X-Y labels

```python
plot_xy([1,2,3,4,5],[3,4,5,6,7], xlabel="X", ylabel="Y", title="Plot X-Y")
```

### Change linewidths, markers, add labels etc. smiilarly to matplotlib

```python
plot_xy([3,4,5],[6,7,10], label="Data 2", lw=4, marker='o', ms=15, ls="--")

"""For brevity you can also write them as args"""
plot_xy([3,4,5],[1,4,7], "ro--", label="Data 3", new_fig=False) # Use new_fig=False for plotting on the previous existing figure
```

```python
plot_xy([1,2,3],[5,7,10], new_fig=False, lw=3, title='Plot X-Y')
```
